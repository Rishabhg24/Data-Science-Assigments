#Theoretical Questions

1. What is unsupervised learning in the context of machine learning?
   - Unsupervised learning is a type of machine learning where the model is trained on data without labeled outputs. In other words, the algorithm is only given the input features (X) but not the corresponding target values (Y).

     The goal of unsupervised learning is to find hidden patterns, structures, or groupings in the data without prior knowledge of the outcomes.

     Key Points:

     * No labels provided: The algorithm doesn’t know the “correct answer.”

     * Focuses on structure: It groups, compresses, or represents data in a meaningful way.

     Common techniques:

     * Clustering (e.g., K-Means, Hierarchical Clustering) → groups similar data points.

      * Dimensionality Reduction (e.g., PCA, t-SNE) → reduces data complexity while preserving important information.

       * Association Rule Learning (e.g., Apriori, Eclat) → finds relationships between variables.

2. How does K-Means clustering algorithm work?
   - The **K-Means clustering algorithm** is an unsupervised learning method used to group data into a predefined number of clusters, denoted by *K*. It works by first randomly selecting *K* initial centroids, which act as the centers of the clusters. Each data point is then assigned to the cluster whose centroid is closest, usually based on Euclidean distance. After assignment, the centroids are updated by calculating the mean of all the points in each cluster, forming new cluster centers. This process of assigning points and updating centroids repeats until the centroids stabilize or a maximum number of iterations is reached. The result is *K* clusters where data points within a cluster are more similar to each other than to points in other clusters.


3. Explain the concept of a dendrogram in hierarchical clustering?
   - A dendrogram is a tree-like diagram used in hierarchical clustering to visually represent how data points or clusters are merged step by step. It starts with each data point treated as its own cluster at the bottom and then progressively merges the closest clusters based on a chosen distance metric (like Euclidean distance) and linkage method (such as single, complete, or average linkage). The vertical axis of the dendrogram represents the distance or dissimilarity at which clusters are joined, while the horizontal axis lists the individual data points. By drawing a horizontal “cut” across the dendrogram at a chosen height, you can decide the number of clusters to form. This makes dendrograms a powerful tool for both understanding the structure of data and determining the optimal number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
   - The main difference between K-Means and Hierarchical Clustering lies in how they form clusters and how the number of clusters is determined. K-Means is a partitioning method that requires the number of clusters (K) to be specified in advance, and it works by iteratively assigning points to the nearest centroid and updating centroids until convergence. In contrast, Hierarchical Clustering is a tree-based method that does not require specifying the number of clusters initially; instead, it builds a hierarchy of clusters either by successively merging smaller clusters (agglomerative) or splitting larger ones (divisive), represented visually using a dendrogram. While K-Means is generally faster and better suited for large datasets, Hierarchical Clustering provides a more detailed structure of relationships between data points but can become computationally expensive for very large datasets.

5. What are the advantages of DBSCAN over K-Means?
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means. Unlike K-Means, it does not require specifying the number of clusters in advance, which makes it more flexible when the optimal number of clusters is unknown. DBSCAN can also discover clusters of arbitrary shapes, whereas K-Means tends to form only spherical clusters around centroids. Additionally, DBSCAN is more robust to outliers, as it treats low-density points as noise rather than forcing them into clusters. These strengths make DBSCAN especially useful for real-world datasets that are noisy, contain irregularly shaped clusters, or where the number of clusters is not predefined.


6.  When would you use Silhouette Score in clustering?
    - The **Silhouette Score** is used in clustering to measure how well data points are grouped within their clusters compared to other clusters. It is applied when you want to **evaluate the quality of clustering results** or **determine the optimal number of clusters**. The score ranges from **-1 to +1**, where a high value (close to +1) means points are well-matched to their own cluster and poorly matched to neighboring clusters, a value near 0 indicates overlapping clusters, and negative values suggest points may be in the wrong cluster. In practice, the Silhouette Score is especially useful for comparing different clustering algorithms (like K-Means, DBSCAN, or Hierarchical Clustering) or different values of *K* to identify which produces the most meaningful separation of clusters.


7.  What are the limitations of Hierarchical Clustering?
    - Hierarchical Clustering, while useful for understanding data structure, has several limitations. It is **computationally expensive**, especially for large datasets, since it requires calculating and storing pairwise distances between all data points, making it unsuitable for very big data. Another drawback is that once a merge or split decision is made, it **cannot be undone**, which may lead to suboptimal clustering results if an early step is incorrect. It is also **sensitive to noise and outliers**, as these can distort the merging process and affect the overall dendrogram. Additionally, the choice of **distance metric and linkage method** (such as single, complete, or average linkage) can significantly change the clustering outcome, making it less robust and sometimes harder to interpret compared to other clustering methods.


8. Why is feature scaling important in clustering algorithms like K-Means?
   - Feature scaling is important in clustering algorithms like **K-Means** because the algorithm relies on **distance calculations** (typically Euclidean distance) to assign points to clusters. If features are on very different scales—for example, "income" measured in thousands and "age" measured in years—the feature with the larger numeric range will dominate the distance calculation, leading to biased clustering results. By applying scaling techniques such as **standardization (z-score normalization)** or **min-max normalization**, all features are brought to a comparable scale, ensuring that each attribute contributes equally to the clustering process. Without scaling, the algorithm may form distorted clusters that reflect differences in scale rather than meaningful patterns in the data.


9. How does DBSCAN identify noise points?
   - DBSCAN identifies noise points by checking whether a data point meets the density requirements defined by two parameters: **ε (epsilon)**, which sets the maximum distance to consider neighbors, and **MinPts**, the minimum number of points required to form a dense region. A point is classified as a **core point** if it has at least MinPts neighbors within ε, a **border point** if it has fewer than MinPts neighbors but lies within the neighborhood of a core point, and a **noise point** (or outlier) if it is neither a core point nor a border point. In other words, noise points are those that do not belong to any dense cluster because they are too far from other points and cannot satisfy the density conditions.


10. Define inertia in the context of K-Means?
    - In the context of **K-Means clustering**, **inertia** refers to the measure of how internally coherent the clusters are, calculated as the **sum of squared distances between each data point and its assigned cluster centroid**. It essentially quantifies how close the data points within a cluster are to their centroid, with lower values indicating tighter, more compact clusters. While inertia is commonly used to evaluate and compare clustering results, it has limitations because it always decreases as the number of clusters increases, which is why it is often combined with methods like the **Elbow Method** to help determine the optimal number of clusters.


11. What is the elbow method in K-Means clustering?
    - The **Elbow Method** in K-Means clustering is a technique used to determine the **optimal number of clusters (K)**. It works by running K-Means with different values of K and calculating the **inertia** (sum of squared distances of points to their nearest centroid) for each value. As K increases, inertia decreases because points are assigned to clusters that better fit them. However, after a certain point, the improvement in inertia becomes marginal, and the curve of inertia vs. K starts to flatten out. The point on the curve where this transition occurs forms an “elbow,” which is considered the best balance between having too few clusters (high inertia, poor fit) and too many clusters (overfitting). This “elbow point” is taken as the optimal number of clusters.


12.  Describe the concept of "density" in DBSCAN?
     - In **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**, the concept of **density** is central to how clusters are formed. Density is defined based on two parameters: **ε (epsilon)**, which specifies the radius of a neighborhood around a point, and **MinPts**, the minimum number of points required within that neighborhood. A region is considered dense if a point has at least MinPts neighbors within its ε-radius, making it a **core point**. Clusters are then built by connecting core points that are within each other’s neighborhoods, while points that do not meet this density requirement are treated as **noise or outliers**. This density-based approach allows DBSCAN to discover clusters of arbitrary shape and effectively separate dense regions from sparse ones.


13.  Can hierarchical clustering be used on categorical data?
     - Yes, **hierarchical clustering** can be used on categorical data, but it requires choosing an appropriate **distance (or similarity) measure** that works with categorical variables. Unlike numerical data, where Euclidean distance is commonly used, categorical data relies on measures such as **Hamming distance, Jaccard similarity, or simple matching coefficient** to capture how similar or different two data points are based on their categories. Once a suitable distance metric is applied, the hierarchical clustering process (agglomerative or divisive) can be carried out in the same way, and the results can be visualized with a **dendrogram**. However, the quality of the clusters heavily depends on the chosen similarity measure, making this step critical when working with categorical data.


14. What does a negative Silhouette Score indicate?
    - A **negative Silhouette Score** indicates that, on average, data points may have been assigned to the **wrong cluster**. This happens when the average distance between a point and the points in its assigned cluster is **greater than** the distance to points in a neighboring cluster. In simpler terms, the point is closer to another cluster than to the one it currently belongs to, suggesting poor clustering quality. A negative score usually means that the clusters are overlapping, not well separated, or that the chosen number of clusters (K) is not appropriate for the dataset.


15. Explain the term "linkage criteria" in hierarchical clustering?
    - In **hierarchical clustering**, the term **linkage criteria** refers to the method used to determine the distance between two clusters when deciding which clusters to merge at each step. Since clusters are groups of multiple points, different strategies are needed to measure inter-cluster distance. Common linkage methods include **single linkage**, which uses the minimum distance between any two points from different clusters; **complete linkage**, which uses the maximum distance; **average linkage**, which considers the average distance between all pairs of points; and **Ward’s method**, which minimizes the increase in variance when clusters are merged. The choice of linkage criteria directly affects the shape and structure of the resulting dendrogram and can lead to different clustering outcomes for the same dataset.


16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
    - K-Means clustering may perform poorly on data with **varying cluster sizes or densities** because it assumes that clusters are **spherical, evenly sized, and equally dense**. The algorithm assigns points to the nearest centroid based on Euclidean distance, which works well when clusters are compact and similar in scale. However, if one cluster is much larger, denser, or differently shaped than another, K-Means may incorrectly split a large cluster into multiple smaller ones or merge smaller clusters together. Additionally, K-Means is sensitive to outliers, which can further distort centroid positions. As a result, it struggles to capture the true structure of datasets with irregular cluster shapes, sizes, or densities.


17.  What are the core parameters in DBSCAN, and how do they influence clustering?
     - The core parameters in **DBSCAN** are **ε (epsilon)** and **MinPts (minimum points)**, and together they define how clusters are formed. **ε** specifies the radius of a neighborhood around a point, determining how close other points must be to be considered neighbors. **MinPts** sets the minimum number of points required within an ε-neighborhood for a point to be classified as a **core point**. If a point has at least MinPts neighbors within ε, it forms the basis of a dense region and can expand a cluster. Points that do not meet this condition but fall within the neighborhood of a core point are considered **border points**, while those that fail both conditions are treated as **noise**. Choosing the right values for ε and MinPts is critical, as small ε values may lead to many small clusters or noise, while large ε values may merge distinct clusters into one.


18.  How does K-Means++ improve upon standard K-Means initialization?
     - **K-Means++** improves upon standard K-Means initialization by selecting the initial centroids in a smarter, more systematic way to avoid poor clustering results. In standard K-Means, centroids are chosen randomly, which can lead to slow convergence or suboptimal clusters if the initial points are poorly placed. K-Means++ reduces this risk by first selecting one centroid randomly, then choosing subsequent centroids with a probability proportional to the square of their distance from the nearest existing centroid. This ensures that centroids are spread out and not clustered too closely together. As a result, K-Means++ often produces better clustering performance, faster convergence, and more consistent results compared to random initialization.


19. What is agglomerative clustering?
    - **Agglomerative clustering** is a type of **hierarchical clustering** that builds clusters in a **bottom-up** manner. It starts with each data point treated as its own individual cluster and then repeatedly merges the two closest clusters based on a chosen **distance metric** (such as Euclidean distance) and **linkage criteria** (like single, complete, or average linkage). This merging process continues step by step until all points are combined into a single cluster or until a stopping condition, such as a predefined number of clusters, is reached. The results are often visualized using a **dendrogram**, which shows the sequence of merges and helps decide the optimal number of clusters. Agglomerative clustering is widely used because it is simple and provides a clear hierarchical structure of the data.


20. What makes Silhouette Score a better metric than just inertia for model evaluation?
    - The **Silhouette Score** is often considered a better metric than just **inertia** for evaluating clustering models because it takes into account both **cohesion** (how close points are to their own cluster) and **separation** (how far points are from other clusters), rather than focusing solely on internal compactness like inertia. While inertia measures the sum of squared distances within clusters, it always decreases as the number of clusters increases, which can be misleading and may encourage overfitting. In contrast, the Silhouette Score ranges from -1 to 1, with higher values indicating well-separated, well-formed clusters, making it more informative for assessing clustering quality and comparing different numbers of clusters or algorithms.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 4 centers
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Scatter plot visualization
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-Means Clustering with 4 Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
    

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Evaluate silhouette scores for k = 2 to 5
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Plot silhouette scores as a bar chart
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k in K-Means')
plt.xticks(k_values)
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# Fit K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Create mesh grid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict cluster for each point on the mesh grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Define color map
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ['#FF0000', '#00FF00', '#0000FF']

# Plot decision boundaries
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.5)

# Plot data points with cluster labels
for i, color in enumerate(cmap_bold):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=color, label=f'Cluster {i}', edgecolor='k', s=50)

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='yellow', marker='X', edgecolor='k', label='Centers')

plt.title('K-Means Clustering with Decision Boundaries (Overlapping Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

36.  Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding

43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side.

```python
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Generate noisy circular data
X, y = make_circles(n_samples=300, noise=0.07, factor=0.5, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.15, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plotting side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot KMeans results
axes[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=50)
axes[0].set_title('KMeans Clustering')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')

# Plot DBSCAN results
# Mark outliers separately (label == -1)
outliers = dbscan_labels == -1
clusters = dbscan_labels != -1
axes[1].scatter(X[clusters, 0], X[clusters, 1], c=dbscan_labels[clusters], cmap='viridis', s=50)
axes[1].scatter(X[outliers, 0], X[outliers, 1], c='red', s=50, label='Outliers')
axes[1].set_title('DBSCAN Clustering')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')
axes[1].legend()

plt.suptitle('Clustering Results on Noisy Circular Data')
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Calculate silhouette coefficient for each sample
silhouette_vals = silhouette_samples(X, labels)

# Sort the silhouette values along with labels for plotting
y_ticks = []
y_lower = 0
plt.figure(figsize=(8, 6))

for i in range(3):
    # Aggregate silhouette scores for samples in cluster i
    ith_cluster_silhouette_vals = silhouette_vals[labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.barh(range(y_lower, y_upper), ith_cluster_silhouette_vals, edgecolor='none', height=1)
    y_ticks.append((y_lower + y_upper) / 2)
    y_lower = y_upper

plt.yticks(y_ticks, [f'Cluster {i}' for i in range(3)])
plt.xlabel('Silhouette Coefficient')
plt.title('Silhouette Coefficients for Each Sample (Iris Dataset)')
plt.axvline(x=np.mean(silhouette_vals), color='red', linestyle='--', label='Average Silhouette Score')
plt.legend()
plt.show()
```
